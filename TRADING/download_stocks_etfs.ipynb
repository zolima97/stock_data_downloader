```python
where_I_arrived = 10000

stock_tik  = stocks[3:]["Yahoo Stock Tickers"].reset_index(drop = True)
sub_stock_tik = stock_tik[where_I_arrived:].reset_index(drop = True)
sub_stock_tik
```

```python
start_time = datetime.now()

stock_list = []
stock_dat = pd.DataFrame(stock_list)
counter = 0

#dichiaro che sub_stock = stock_tik
stock_tik = sub_stock_tik
for indice, elem in enumerate(stock_tik[:20000]):
    if (indice+1 < len(stock_tik[:20000]) and indice - 1 >= 0):
        prev_el = str(stock_tik[:20000][indice-1])
        curr_el = str(elem)
        stock_df = yf.download(curr_el, 
                          start='2000-01-01', 
                          end=today, 
                          progress=False)
        
        stock_open = pd.DataFrame(stock_df["Close"])
        #print(curr_el)
        stock_dat = pd.merge(stock_dat, stock_open, left_index=True, right_index=True, how = "outer", suffixes=('_'+ prev_el, '_'+ curr_el))
        counter += 1
        
        if counter % 100 == 0:
            print("we reached the number:", counter)

stock_dat = stock_dat.dropna(how = "all", axis = 1)
        


end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

```python
stock_dat_10000_30000 = stock_dat.to_parquet('stock_dat_10000_30000.parquet', compression='GZIP')
```
